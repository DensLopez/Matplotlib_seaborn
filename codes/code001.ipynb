````python
import matplotlib.pyplot as plt
import numpy as np
````

Una vez se importa, procedemos a trabajar con las variables.

````python
#Creando una variable
x = np.linspace(0,5,10)
y = x**2

#Visualización de variables
print(f'X = {x}\nTipo = {type(x)}\n\nY = x^2 = {y}\nTipo = {type(y)}\n')
````

````python
#Trabajando con Pyplot

#Generando gráfica
plt.plot(x,y)
#Mostrando gráfica
plt.show()
````

## Cambiando los colores de la gráfica
------------
Podemos cambiar los colores de la gráfica mediante el comando:
```python
plt.plot(x,y,'character')
#Ejemplo
plt.plot(x,y,'r')   #Nos muestra en color rojo

```

|character|	color|
|---------|---------|
|‘b’|	blue|
|‘g’|	green|
|‘m’|	magenta|
|'r'| red|
|'c'| cyan|
|'y'| yellow|
|'k'| black|
|'w'| white|



````python
plt.plot(x,y,'c')
plt.show
````

##### Aplicando en código
```python
plt.plot(x,y, 'yD:') #grafica de color amarillo, con diamantes y puntos consecutivos
plt.show()
```

````python
plt.plot(x,y,'yD:')
plt.show
````

````python
#Rojo-Squares-Continuidad
plt.plot(x,y,'rs-')
#Graficando
plt.show
````

````python
#Graficando histograma con 2 variables
plt.hist(x,y)
plt.show()
````

# Nota:
Una buena práctica es que cuando estés haciendo histogramas, siempre explora multiples anchos de bins
<ol type="A">
  <li>bins = 1</li>
  <li>bins = 3</li>
  <li>bins = 5</li>
  <li>bins = 15</li>

````python
#Graficando histograma con 1 variable
#Con los 10 datos de X
plt.hist(x)
plt.show()
````

````python
#Graficando gráfica tipo pastel
#Con los 10 datos de X
plt.pie(x)
plt.show()
````

````python
#Grafica de relacion (Scatter)
plt.scatter(x,y)
plt.show
````

````python
#Ver la distribución de los datos
plt.boxplot(x)
plt.show()
````
