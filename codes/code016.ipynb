# Heatmap
Heatmap es un tipo de gráfico enfocado a una estructura matricial. Heatmap correlaciona todas las variables numéricas del dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
#Vamos a llevar tips a 
#una estructura matricial
tips.corr(numeric_only=True)  
```

### Explicación 
La correlación va de -1 a 1:
- (-1): Correlación totalmente indirecta.
- (0): No existe correlación.
- (+1): Es una correlación directa.
Por supuesto los valores intermedios van a hacer referencia al tipo de correlación del que estamos hablando (negativa, neutra o postiva).

Por ejemplo en la tabla anterior la diagonal principal es una correlación directa y las combinaciones de las demas me dan otros resultados.

De manera gráfica veremos esta correlación con **HEATMAP**

## Heatmap

```python
#Graficando
sns.heatmap(tips.corr(numeric_only=True) )
plt.show()
```

Es muy útil poder darnos cuenta de manera gráfica como se relacionan las variables y con una escala de color, podemos darle mayor contexto con más parametros.

```python
#heatmap
sns.heatmap(tips.corr(numeric_only=True),annot=True )
plt.show()
```

Podemos decir que la ``propina`` está relacionada con `la cuenta total` en un ``68%``|, es decir lo que vimos antes en el scatter plot, ahora lo puedo ver gráficamente.

### cmap = 

```python
#heatmap
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.show()
```

Entre más tiende al **rojo** es más `calido`, entre más tiende al **azul** es más `frío`. En la documentación se encuentran los diferentes mapas de colores con los que podemos trabajar

### linewidths, linecolor

```python
#heatmap
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor='green')
plt.show()
```

### vmin=0 y vmax=1

```python
#heatmap
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor='green',vmin=0,vmax=1)
plt.show()
```

### Valores default heatmap

```python
#heatmap
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor='green')
plt.show()
```

### Quitando barra cbar=False

```python
#heatmap
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor='green',cbar=False)
plt.show()
```

### Lo mejor es trabajar con barra para dar contexto

```python
#heatmap
sns.heatmap(tips.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor='green')
plt.show()
```
