# Set
Set es un método de Seaborn que permite configurar el tema; es decir el estilo, fuente y color de las gráficas.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
````

````python
#Haciendo grafica de barras
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
#Seaborn esta montado sobre matplotlib
````

````python
#Definiendo un estilo
sns.set_theme(style='dark',palette='deep')
#Haciendo grafica de barras
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
````

## Estilos:
styledict, or one of 
`{darkgrid, whitegrid, dark, white, ticks}`

## Paletas

````python
for index,paleta in enumerate(sns.palettes.SEABORN_PALETTES.keys()):
    print(f'paleta {index+1} = {paleta}')
````

````python
#Definiendo un estilo
sns.set_theme(style='ticks',palette='muted')
#Haciendo grafica de barras
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
````

````python
#Definiendo un estilo
sns.set_theme(style='whitegrid',palette='pastel')
#Haciendo grafica de barras
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
````

## Cambiando tipografia y escala

````python
#Definiendo un estilo
sns.set_theme(style='whitegrid',palette='pastel',font='Ink Free',font_scale=1)
#Haciendo grafica de barras
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
````

## Solución a problemas de fuente
Solución a problema de fuente en VS CODE:

Si tienes fuentes en Windows y quieres traer a un entorno de WSL

Primero copia las fuentes de la ruta de Windows:
``C:\Windows\Fonts``

Despues pega las fuentes en el siguiente directorio:<br>
Si tienes Linux: con Miniforge:
``\home\tu_usuario\miniforge3\lib\python3.12\site-packages\matplotlib\mpldata\fonts\ttf``

Si tienes Linux con anaconda3:
``/home/***/anaconda3/pkgs/matplotlib-base-3.1.3-py37hef1b27d_0/lib/python3.7/sitepackages/matplotlib/mpl-data/fonts/ttf/``

Y corre este comando en terminal 
```bash
rm ~/.cache/matplotlib -rf 
```

NOTA: Es importante que no estes ejecutando el VS CODE.
