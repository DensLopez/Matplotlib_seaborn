# Bar plot
Este tipo de gráfica nos permite mostrar variables categoricas o que contienen texto, lo que en la ciencia de datos es importante para la visualización de datos. <br>
Matplotlib nos ofrece ciertas funciones para este tipo de gráficas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Definiendo variables
paises = ['In','Ja','Méx','Col','Ale']
poblacion = [1000,800,500,700,285]
```

```python
#Haciendo gráficas de barras
plt.bar(paises,poblacion)
```

```python
#Personalizando
#Haciendo gráficas de barras
plt.bar(paises,poblacion,width=0.5, color='red')
```

```python
#Alternando colores
#Haciendo gráficas de barras
plt.bar(paises,poblacion,width=0.5, color=['red','blue'])
```

## Modificando parametros:
Supongamos que no puedo acceder a las variables categoricas y simplemente entro a una de las propiedades:
- `plt.xticks`: Configuraciones del eje x
  - `np.arange(5)`: que me de una lista o posiciones del arreglo y así poder entrar al indice del eje x

```python
#Alternando colores
#Haciendo gráficas de barras
#Cambiando configuraciones de eje x
plt.xticks(np.arange(5),('India','Japón','Mexico','Colombia','Alemania'))
plt.bar(paises,poblacion,width=0.5, color=['red','blue','c','yellow','gray'])
```

### Rotando para evitar el solapamiento

```python
#Alternando colores
#Cambiando configuraciones de eje x
plt.xticks(np.arange(5),('India','Japón','Mexico','Colombia','Alemania'),rotation=45)
plt.bar(paises,poblacion,width=0.5, color=['red','blue','c','green','gray'])
```

# Graficando barras horizontales

```python
#Alternando colores
#Haciendo gráficas de barras
#Cambiando configuraciones de eje x
plt.barh(paises,poblacion, color=['red','blue','c','green','gray'])
plt.yticks(np.arange(5),('India','Japón','Mexico','Colombia','Alemania'),rotation=45)
```
