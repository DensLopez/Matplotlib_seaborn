# Otro tipos de gráfico

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## Histograma

```python
#Analizar distribución y frecuencia
#histograma
plt.hist(data)
#bins = 10 por default
plt.show()
```

#### Bins = Número de barras en la cuál va a llegar la distribución de datos

```python
#Jugar cn una variable importante 
#Bins en histogramas
plt.hist(data,bins=50)
plt.show()
```

```python
#Modificando parametros
plt.hist(data,bins=15,histtype='bar')
plt.show()
```

```python
#Modificando parametros
plt.hist(data,bins=15,histtype='step')
plt.show()
```

## Boxplot
Grafico de cajas y bigotes

```python
#Boxplot
plt.boxplot(data)
plt.show()
```

```python
#Explorando propiedades
plt.boxplot(data,vert=False)
plt.show()
```

```python
#Explorando propiedades
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
#Notch me dice como se encuentran los datos en distribuación 
plt.show()
```

```python
#Agregando data de tipo outlayer (con anomalias)
data=np.append(data,200)
#Explorando propiedades
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
#Notch me dice como se encuentran los datos en distribuación 
plt.show()
```

Aquí toma en cuenta el data anomalo. Podemos quitarlo con la sentencia:<br>
`showfliers=False`


```python
#Agregando data de tipo outlayer (con anomalias)
data=np.append(data,200)
#Explorando propiedades
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
#Notch me dice como se encuentran los datos en distribuación
plt.show() 
```

#### Si le quito el show out layers
Me muestra todos los datos con los que estoy trabajando.<br>
`showfliers=True`

```python
#Agregando data de tipo outlayer (con anomalias)
data=np.append(data,200)
#Explorando propiedades
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=True)
#Notch me dice como se encuentran los datos en distribuación 
plt.show()
```

## Scatterplot (Dispersión)

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30*np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
#Graficando Scatter ()
plt.scatter(x,y)
```

```python
#Graficando Scatter ()
#Agregando más variable
plt.scatter(x,y,s=area)
```

## Explicación:
El tamaño de las burbujas va a tener referencia a una 3ra variable que es: area 
¿Qué pasa si agrego otra variable?


```python
#Graficando Scatter ()
#Agregando más variable
plt.scatter(x,y,s=area,c=colors)
```

Con esto de los colores me podría servir para hacer una clasificación de lo que estoy haciendo.

```python
#Cambiando markers
#Graficando Scatter ()
plt.scatter(x,y,s=area,c=colors,marker='o')
```

|Marker|Tipo|
|---|---|
|'x'|x|
|'p'|poligono|
|'s'|Cuadrado|
|'d'|Diamante|
|'^'|Triangulo hacia arriba|
|'v'|Triangulo hacia abajo|
|'<'|Triangulo hacia la izquierda| 
|'>'|Triangulo hacia la derecha|
|'o'|Circulo|
|'\|'|Palito|
|'*'|Estrella|
|'+'|Cruz|
|'.'|Puntos pequeños|



#### Ajustando Alpha

```python
#Cambiando markers
#Graficando Scatter ()
plt.scatter(x,y,s=area,c=colors,marker='o',alpha=0.5)
plt.show()
```

De esta manera puedo ver si hay puntos o datos sobrepuestos
